```python
import os
import pickle
import tensorflow as tf
from utils import create_model, get_logger
from model import Model
from loader import input_from_line
from train import FLAGS, load_config
from predict import extraction_result
import pandas as pd
```

```python
data = pd.read_excel(r'yq_bufen.xlsx',encoding='utf-8')
```

```python
sentence = (data.sentence).to_list()
```

```python
sentence = list(set(sentence))
print(len(sentence))
person_list = []
org_list = []
other_list = []
```

```python
config = load_config(FLAGS.config_file)
logger = get_logger(FLAGS.log_file)
    # limit GPU memory
tf_config = tf.ConfigProto()
tf_config.gpu_options.allow_growth = True
with open(FLAGS.map_file, "rb") as f:
    tag_to_id, id_to_tag = pickle.load(f)
        
with tf.Session(config=tf_config) as sess:
    model = create_model(sess, Model, FLAGS.ckpt_path, config, logger)
    for i in range(len(sentence)):
        s = sentence[i]
        result = model.evaluate_line(sess, input_from_line(s, FLAGS.max_seq_len, tag_to_id), id_to_tag)
        res = result['entities']
        person = ""
        org = ""
        other = ""
        for di in res:
            if di['type'] == 'person':
                person = person+" "+di['word']
            elif di['type'] == 'org':
                org = org+" "+di['word']
            elif di['type'] == 'other':
                other = other+" "+di['word']
        person_list.append(person.strip())
        org_list.append(org.strip())
        other_list.append(other.strip())
        print(i)
        #print(result['entities'])
```

```python
person_list = list(set(person_list))
org_list = list(set(org_list))
other_list = list(set(other_list))
```
