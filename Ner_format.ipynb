```python
person_list = (data.person).tolist()
org_list = (data.org).tolist()
other_list = (data.other).tolist()
sentence_list = (data.sentence).tolist()
```

转ner格式函数

```python
def trans_t(sentence,tag_list,input_list,flag):
    word_list = []
    if flag == '0':
        f1 = "B-person"
        f2 = "I-person"
        temp = ""
        input_list = input_list.split()
        for word in input_list:
            if word!= 'xxx':
                word_list.append(word)
    elif flag == '1':
        f1 = 'B-org'
        f2 = 'I-org'
        input_list = input_list.split()
        for word in input_list:
            if word!= 'xxx':
                word_list.append(word)
    elif flag == '2':
        f1 = 'B-other'
        f2 = 'I-other'
        input_list = input_list.split()
        for word in input_list:
            if word!= 'xxx':
                word_list.append(word)
    s2 = sentence
    len_sen = len(sentence)
    word_list.sort(key = lambda i:len(i),reverse=True) 
    for word in word_list:
        while(word in s2):
            start = s2.find(word)
            len_word = len(word)#对s1进行标记
            for i in range(len_sen):
                    if i >=start and i < (start+len_word):
                        if i == start: 
                            tag_list[i] = f1
                            #print("x")
                        else : 
                            tag_list[i] = f2
                            #print("xx")
            temp = ""
            for i in range(len_word):temp += "y"
            s2 = s2.replace(word,temp,1)
           
    return tag_list,s2
```

```python
write_file = "yq_entity.test2000"
for i in range(2500,3000):
    person = person_list[i]
    org = org_list[i]
    other = other_list[i]
    if isinstance(sentence_list[i],int):
        continue
    sentence = sentence_list[i].strip().replace(' ','')
    sentence = "".join(sentence.split())
    len_sen = len(sentence)
    print(i)
    tag_list = []
    for x in  range(len_sen): tag_list.append('O')
    sen_tag,s2 = trans_t(sentence,tag_list,person,'0')
    sen_tag,s2 = trans_t(s2,sen_tag,org,'1')
    sen_tag,s2 = trans_t(s2,sen_tag,other,'2')
    with open(write_file,"a+",encoding='utf-8') as f:
        for t in range(len(sentence)):
            #print(sentence[t]+" "+sen_tag[t])
            f.write(sentence[t]+" "+sen_tag[t]+"\n")
        f.write("\n")
```
